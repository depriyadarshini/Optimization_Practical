```python
"""XYZ Clothing is introducing a new line. The marketing department is going to invest social spending into facebook ads
for two groups: 1) Men and 2) women. The cost per click on the previous ad campaign was $1.46 for women and $1.64 for men.
For every 49 clicks 1 woman converts versus 1 man for every 26 clicks. We having the following goals:
1. get 400 conversions from women
2. get 500 conversions from men
3. Limit the expense of marketing to 100,000
Assume goals are all equally important
"""
```

```python
from pulp import *
model = pulp.LpProblem("under/overachievement minimizing problem", pulp.LpMinimize)
x1 = pulp.LpVariable("x1", lowBound=0, cat='Integer') #number of women who see
y1 = pulp.LpVariable("y1", lowBound=0, cat='Integer') #number of men who see
x2 = pulp.LpVariable('x2', lowBound=0, cat='Continuous')#cost
d1_plus = pulp.LpVariable('d1_plus', lowBound=0, cat='Integer')#over rep group one
d1_neg = pulp.LpVariable("d1_neg", lowBound=0, cat='Integer') #under represent group one
d2_plus = pulp.LpVariable('d2_plus', lowBound=0, cat='Integer')#ober rep group 2
d2_neg = pulp.LpVariable('d2_neg', lowBound=0, cat='Integer')#under represent group 2
d3_plus = pulp.LpVariable('d3_plus', lowBound=0, cat='Continuous')#over expense
d3_neg = pulp.LpVariable('d3_neg', lowBound=0, cat='Continuous')#under expense


# Objective function
model += d1_neg + d2_neg + d3_plus, "under_over"


# Constraints
model += x1 + d1_neg - d1_plus == 3711512 #number of women to see to get 300 women to convert
model += y1 + d2_neg - d2_plus == 2460080 #number men to see to get 300 men to convert
model += x1*0.0029 + y1*0.0022 - d3_plus + d3_neg == 18000

model.solve()
LpStatus[model.status]

print('number of group one individuals contacted ={}'.format(x1.varValue))
print('number of group two individuals contacted = {}'.format(y1.varValue))
print('Overachievement of letters sent to group 1 ={}'.format(d1_plus.varValue))
print('Overachievement of letters sent to group 2 = {}'.format(d2_plus.varValue))
print('Over budget = {}'.format(d3_plus.varValue))
print('Under budget = {}'.format(d3_neg.varValue))
```
